```python
def channel_details(c_id):
  request = youtube.channels().list(
        part="snippet,contentDetails,statistics",
        id=c_id
    )

  response = request.execute()

  z = dict(cid = response['items'][0]['id'],
      ti = response['items'][0]['snippet']['title'],
      d = response['items'][0]['snippet']['description'],
      v = response['items'][0]['statistics']['viewCount'])

  return z
```

```python
def playlist_details(c_id):
  request = youtube.playlists().list(
        part="snippet,contentDetails",
        channelId=c_id,
        maxResults=25
    )
  response = request.execute()

  pz = dict(pid = response['items'][0]['id'],
       pc = response['items'][0]['snippet']['channelId'],
       pd= response['items'][0]['snippet']['channelTitle'])

  return pz
```

```python
def video_details(v_id):  #Ks-_Mh1QhMc
  request = youtube.videos().list(
        part="snippet,contentDetails,statistics",
        id=v_id
    )
  response = request.execute()

  vd = dict(vid = response['items'][0]['id'],
       vcid = response['items'][0]['snippet']['channelId'],
       vdesc = response['items'][0]['snippet']['description'],
       vdt = response['items'][0]['snippet']['publishedAt'],
       vli = response['items'][0]['statistics']['likeCount'],
       vcm = response['items'][0]['statistics']['commentCount'],
       vvw = response['items'][0]['statistics']['viewCount'],
       vf = response['items'][0]['statistics']['favoriteCount'],
       vt =  response['items'][0]['snippet']['thumbnails'],
       vdur =   response['items'][0]['contentDetails']['duration']
       )

  return vd
```
